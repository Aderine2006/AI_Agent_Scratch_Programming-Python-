```python
class Planner:
  #Creating an AI Agent FOR Planner , Executor , Critic
  def plan(self,goal):
    return ["Collect data","Analyze data","Generate report"]
class Executor:
  def execute(self,task):
    if task == "Collect data":
      return "Data collected"
    elif task == "Analyze data":
      return "Analysis incomplete"
    elif task == "Generate report":
      return "Report generated"
class Critic:
  def critique(self,task,result):
    if "incomplete" in result:
      return False
    return True

planner = Planner()
executor = Executor()
critic = Critic()

goal = "Data Analysis Project"
tasks = planner.plan(goal)
for task in tasks:
  result = executor.execute(task)
  approved = critic.critique(task,result)
  if not approved:
    print(f"Critic: Re-doing {task}")
    result = executor.execute(task.replace("Analyze", "Re-analyze"))
  print(f"Final result for {task}: {result}")
```
