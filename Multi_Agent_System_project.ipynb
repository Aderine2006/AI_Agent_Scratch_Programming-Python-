```python
#-------------------------------------------------#
#Multi-Agent System Project
#Samrt Traffic Signal Control System
#-------------------------------------------------#

import random

#------------------VECHILE DENSITY AGENT ---------#

class VechileDensityAgent:
  def sense_density(self):

    #Random vechicle count simulation
    density = random.randint(5, 60)
    print(f"[Density Agent] Vechicle Count: {density}")
    return density

# -----------------EMERGENCY AGENT------------------#

class EmergencyAgent:
  def detect_emergency(self):

    #Random Emergency detection
    emergency = random.choice([True, False, False, False])
    print(f"[Emergency Agent] Emergency Detected: {emergency}")
    return emergency

class TrafficSignalAgent:
  def set_signal(self,green_time, priority = False):
    if priority:
      print(f"[Signal Agent] EMERGENCY PRIORITY: GREEN TIME: {green_time}s")
    else:
      print(f"[Signal Agent] GREEN TIME: {green_time}s")

#----------------------COORDINATOR AGENT -------------------------

class CoordinatorAgent:
  def decide(self, density, emergency):
    if emergency:
          return 60,True
    if density > 40:
          return 50,False
    elif density > 20:
          return 35,False
    else:
          return 30,False

#-------------------- MAIN HAS SYSTEM ----------------
def main():
  print("\nSMART TRAFFIC SIGNAL SYSTEM {HAS}")
  print("-------------------------------------------------")

  density_agent = VechileDensityAgent()
  emergency_agent = EmergencyAgent()
  signal_agent = TrafficSignalAgent()
  coordinator = CoordinatorAgent()

  for cycle in range(1,6):
    print(f"\n---Traffic Cycle {cycle}---")
    density = density_agent.sense_density()
    emergency = emergency_agent.detect_emergency()
    green_time, priority = coordinator.decide(density, emergency)
    signal_agent.set_signal(green_time, priority)
if __name__ == "__main__":
  main()
```
